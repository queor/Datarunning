##### 문자열 메서드

count()
문자열 내부에 포함된 특정 문자열의 개수를 반환하는 메소드  

인덱스를 지정해 검색 범위를 지정할 수도 있음

```python
print('apple watch is very expensive'.count('a'))
print('apple watch is very expensive'.count('apple'))
print('apple watch is very expensive'.count('samsung'))
```

```python
# 문자열 메서드 동작 방식은 단순 매칭 방식으로 사용 시 주의
print("apple's application is a very good app".count('app'))
```

find()
문자열 내부에 포함된 특정 문자열을 찾고자 할 때 사용  
찾고자하는 문자열이 있으면 그 문자열이 처음으로 나온 위치(인덱스)를 반환  
만일 찾고자 하는 문자열이 없는 경우 -1을 반환

인덱스를 이용해 검색 범위 지정 가능  
지정하지 않으면 처음부터 찾고,  
시작할 인덱스를 지정하는 경우에도 지정된 인덱스부터 찾음

rfind()메소드를 통해 오른쪽에서부터 값을 탐색 가능

```python
print('apple wath is very expensive'.find('a'))
print('apple wath is very expensive'.find('apple'))
print('apple wath is very expensive'.find('samsung'))
```

```python
# 안녕하세요 여러분 반갑습니다. 오늘은 빅데이터 첫 수업 시간입니다.
# "빅데이터"라는 단어가 존재할 시 존재, 없으면 미존재라는 단어가 출력되는 코드 작성
string = '빅데이터 안녕하세요 여러분 반갑습니다. 오늘은 빅데이터 첫 수업 시간입니다.'

if string.find("빅데이터") == -1:
    print('미존재')
else:
    print('존재')
```

대소문자 변환

|메소드|의미|
|------|----|
|upper()|모두 대문자로 변환한 결과를 반환|
|lower()|모두 소문자로 변환한 결과를 반환|
|capitalize()|첫 글자는 대문자로 변환하고 나머지 문자는 소문자로 변환한 결과 반환|

```python
print('apple'.upper())
print('apPle'.upper())
print('apple'.lower())
print('apple'.lower())
print('APPLE'.capitalize())
print('APPLE'.capitalize())
print('Apple is very expensive'.capitalize())
```

join()
인수로 전달한 반복가능객체(문자열, 리스트 등)의 각 요소 사이에  
문자열을 포함시켜 새로운 문자열을 만들고 그 결과를 반환

```python
# 권장되지 않는 방식
'안녕하세요' + ' 여러분'
```

```python
# 권장하는 방식
' '.join(['안녕하세요', '여러분', '오늘은', '빅데이터', '첫', '수업', '시간입니다.'])
```

split()

하나의 문자열을 여러 개의 문자열로 분리 후 리스트로 반환하는 메소드  
기본적으로는 공백문자로 분리하지만 특정 문자열을 기준으로 분리 가능

```python
print('안녕하세요 여러분 오늘은 빅데이터 첫 수업 시간입니다.'.split())
print('안녕하세요 여러분 오늘은 빅데이터 첫 수업 시간입니다.'.split('-'))
```

불필요한 문자열 제거 메소드

|메소드|의미|
|--------|--------|
|lstrip  |왼쪽 끝에 있는 공백을 제거 후 반환|
|rstrip  |오른쪽 끝에 있는 공백을 제거 후 반환|
|strip   |양쪽 끝에 있는 공백을 제거 후 반환|

```python
print('안녕하세요 여러분 오늘은 빅데이터 첫 수업 시간입니다.')
print('                  안녕하세요 여러분 오늘은 빅데이터 첫 수업 시간입니다.           '.lstrip())
print('                  안녕하세요 여러분 오늘은 빅데이터 첫 수업 시간입니다.            '.rstrip())
print('                  안녕하세요 여러분 오늘은 빅데이터 첫 수업 시간입니다.            '.strip())
```

문자열 탐색 메소드

|메소드|의미|
|--------|--------|
|startswith  |문자열이 특정 단어로 시작하면 True, 아니면 False|
|endswith    |문자열이 특정 단어로 끝나면 True, 아니면 False|

```python
print('apple is very expensive'.startswith('s'))
print('apple is very expensive'.endswith('e'))
```
